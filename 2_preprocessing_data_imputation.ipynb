```python
data_mipa = pd.read_pickle("DataPreprocessing/data_mipa.pkl")
data_sosial = pd.read_pickle("DataPreprocessing/data_sosial.pkl")

## Data Mipa
data_mipa['JUR'] = data_mipa.NIM.replace(insert, regex=True)


### Change Class 'IPK_TERAKHIR' & 'STATUS_LULUS' to categorical;
data_mipa['IPK_TERAKHIR'] = data_mipa['IPK_TERAKHIR'].apply(partial(ip_tocategory, feat='IPK_'))
data_mipa['STATUS_LULUS'] = data_mipa.apply(lambda row: status_grad(row),axis=1)
display(data_mipa.head())
# data_mipa.to_csv('sssrr.csv')


### Data Sosial
data_sosial['JUR'] = data_sosial.NIM.replace(insert, regex=True)
# data_sosial.to_csv('ssssss.csv')

### Change Class 'IPK_TERAKHIR' & 'STATUS_LULUS' to categorical;
data_sosial['IPK_TERAKHIR'] = data_sosial['IPK_TERAKHIR'].apply(partial(ip_tocategory, feat='IPK_'))
data_sosial['STATUS_LULUS'] = data_sosial.apply(lambda row: status_grad(row),axis=1)
display(data_sosial.head())

print(len(data_mipa))
print(len(data_sosial))
# data_sosial.to_csv('ssssssss.csv')
```
