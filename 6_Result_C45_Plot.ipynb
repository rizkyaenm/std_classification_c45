# Based on Data Preprocessing

```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral11
from bokeh.layouts import row
from bokeh.io import output_file, save 

import pandas as pd


data_mipa_ipk = pd.read_csv('DataResult/original/mipa-ipk/Summary.csv')
data_mipa_lulus = pd.read_csv('DataResult/original/mipa-lulus/Summary.csv')
data_sosial_ipk = pd.read_csv('DataResult/original/sosial-ipk/Summary.csv')
data_sosial_lulus = pd.read_csv('DataResult/original/sosial-lulus/Summary.csv')

data_mipa_ipk_dem = pd.read_csv('DataResult/demografis/mipa-ipk/Summary.csv')
data_mipa_lulus_dem = pd.read_csv('DataResult/demografis/mipa-lulus/Summary.csv')
data_sosial_ipk_dem = pd.read_csv('DataResult/demografis/sosial-ipk/Summary.csv')
data_sosial_lulus_dem = pd.read_csv('DataResult/demografis/sosial-lulus/Summary.csv')

data_mipa_ipk_random = pd.read_csv('DataResult/random/mipa-ipk/Summary.csv')
data_mipa_lulus_random = pd.read_csv('DataResult/random/mipa-lulus/Summary.csv')
data_sosial_ipk_random = pd.read_csv('DataResult/random/sosial-ipk/Summary.csv')
data_sosial_lulus_random = pd.read_csv('DataResult/random/sosial-lulus/Summary.csv')

data_mipa_ipk_over = pd.read_csv('DataResult/oversampling/mipa-ipk/Summary.csv')
data_mipa_lulus_over = pd.read_csv('DataResult/oversampling/mipa-lulus/Summary.csv')
data_sosial_ipk_over = pd.read_csv('DataResult/oversampling/sosial-ipk/Summary.csv')
data_sosial_lulus_over = pd.read_csv('DataResult/oversampling/sosial-lulus/Summary.csv')

output_notebook()

columns = ['Correctly Classified', 'Incorrectly Classified', 'Kappa', 'MEE',
            'RMSE','Relative abs. error', 'Root Relative abs. error']
name = ['Data mipa ipk', 'Data mipa lulus',
       'Data sosial ipk', 'Data sosial lulus']

source_mipa_ipk = ColumnDataSource(data_mipa_ipk)
source_mipa_lulus = ColumnDataSource(data_mipa_lulus)
source_sosial_ipk = ColumnDataSource(data_sosial_ipk)
source_sosial_lulus = ColumnDataSource(data_sosial_lulus)

source_mipa_ipk_dem = ColumnDataSource(data_mipa_ipk_dem)
source_mipa_lulus_dem = ColumnDataSource(data_mipa_lulus_dem)
source_sosial_ipk_dem = ColumnDataSource(data_sosial_ipk_dem)
source_sosial_lulus_dem = ColumnDataSource(data_sosial_lulus_dem)

source_mipa_ipk_over = ColumnDataSource(data_mipa_ipk_over)
source_mipa_lulus_over = ColumnDataSource(data_mipa_lulus_over)
source_sosial_ipk_over = ColumnDataSource(data_sosial_ipk_over)
source_sosial_lulus_over = ColumnDataSource(data_sosial_lulus_over)

source_mipa_ipk_random = ColumnDataSource(data_mipa_ipk_random)
source_mipa_lulus_random = ColumnDataSource(data_mipa_lulus_random)
source_sosial_ipk_random = ColumnDataSource(data_sosial_ipk_random)
source_sosial_lulus_random = ColumnDataSource(data_sosial_lulus_random)

hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("(x,y)", "($x, $y)"),
           
        ]
    )

p=figure(width =800, height =400, tools=[hover],title="Original Data")
p1=figure(width =800, height =400, tools=[hover], title="Data Demografis")
p2=figure(width =800, height =400, tools=[hover], title = "Data Random")
p3=figure(width =800, height =400, tools=[hover], title = "Data Oversampling")


p.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_ipk, 
       line_width=2, color = "red",legend = name[0])
p.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_lulus, 
       line_width=2, color = "purple",legend = name[1])
p.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_ipk,
       line_width=2, color = "green",legend = name[2])
p.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_lulus,
       line_width=2, color = "pink",legend = name[3])


p1.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_ipk_dem, 
       line_width=2, color = "red",legend = name[0])
p1.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_lulus_dem, 
       line_width=2, color = "purple",legend = name[1])
p1.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_ipk_dem,
       line_width=2, color = "green",legend = name[2])
p1.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_lulus_dem,
       line_width=2, color = "pink",legend = name[3])

p2.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_ipk_random, 
       line_width=2, color = "red",legend = name[0])
p2.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_lulus_random, 
       line_width=2, color = "purple",legend = name[1])
p2.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_ipk_random,
       line_width=2, color = "green",legend = name[2])
p2.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_lulus_random,
       line_width=2, color = "pink",legend = name[3])


p3.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_ipk_over, 
       line_width=2, color = "red",legend = name[0])
p3.line(x='Confidence Value', y = 'Correctly Classified', source=source_mipa_lulus_over, 
       line_width=2, color = "purple",legend = name[1])
p3.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_ipk_over,
       line_width=2, color = "green",legend = name[2])
p3.line(x='Confidence Value', y = 'Correctly Classified', source=source_sosial_lulus_over,
       line_width=2, color = "pink",legend = name[3])

p.axis.minor_tick_in = -1
p.axis.minor_tick_out = 6
p.legend.location = "top_right"
p.legend.click_policy="hide"

p1.axis.minor_tick_in = -1
p1.axis.minor_tick_out = 6
p1.legend.location = "top_right"
p1.legend.click_policy="hide"

p2.axis.minor_tick_in = -1
p2.axis.minor_tick_out = 6
p2.legend.location = "top_right"
p2.legend.click_policy="hide"

p3.axis.minor_tick_in = -1
p3.axis.minor_tick_out = 6
p3.legend.location = "top_right"
p3.legend.click_policy="hide"
```
