# Oversampling

## Dataframe to Array

```python


# ### Assigned new variable to contain array of class 
# ### from data mipa and sosial 

class_mipa_ipk = data_mipa_ipk['IPK_TERAKHIR']
class_mipa_lulus = data_mipa_lulus['STATUS_LULUS']

class_sosial_ipk = data_sosial_ipk['IPK_TERAKHIR']
class_sosial_lulus = data_sosial_lulus['STATUS_LULUS']


# ### Assigned new variable to contain array of parameters
# ### from data mipa and sosial 
attr_mipa_ipk = data_mipa_ipk.iloc[:,:10]
attr_mipa_lulus = data_mipa_lulus.iloc[:,:10]

display(attr_mipa_ipk.head())
display(attr_mipa_lulus.head())

attr_sosial_ipk = data_sosial_ipk.iloc[:,:10]
attr_sosial_lulus = data_sosial_lulus.iloc[:,:10]

# display(attr_sosial_ipk.head())
# display(attr_sosial_lulus.head())
```

## Oversampling data mipa

```python
# #### OVERSAMPLING DATA MIPA CLASS IPK

# #### Convert Pandas to numpy 
attr_mipa_ipk = attr_mipa_ipk.reset_index()
attr_mipa_lulus = attr_mipa_lulus.reset_index()

savedcols_ipk = attr_mipa_ipk.columns
savedcols_lulus = attr_mipa_lulus.columns

class_mipa_ipk = class_mipa_ipk.values.astype("U")
class_mipa_lulus = class_mipa_lulus.values.astype("U")

attr_mipa_ipk =attr_mipa_ipk.values.astype("U")
attr_mipa_lulus =attr_mipa_lulus.values.astype("U")


print (type(attr_mipa_ipk))
print (len(attr_mipa_ipk))
print (len(class_mipa_ipk))

print (attr_mipa_ipk)
print (class_mipa_ipk)

# #### oversampling
rus = RandomOverSampler()
data_mipa_ipk_over, class_mipa_ipk_over = rus.fit_sample(attr_mipa_ipk,class_mipa_ipk)
data_mipa_lulus_over, class_mipa_lulus_over = rus.fit_sample(attr_mipa_lulus, class_mipa_lulus)

# # print( data_mipa_ipk_over)
# # print(savedcols)

# # convert back to pandas
data_mipa_ipk_over = pd.DataFrame(data_mipa_ipk_over, columns=savedcols_ipk).set_index("index")
data_mipa_ipk_over['IPK_TERAKHIR'] = class_mipa_ipk_over
data_mipa_ipk_over = data_mipa_ipk_over.reset_index(drop=True)
display(data_mipa_ipk_over.head())

data_mipa_lulus_over = pd.DataFrame(data_mipa_lulus_over, columns=savedcols_lulus).set_index("index")
data_mipa_lulus_over['STATUS_LULUS']=class_mipa_lulus_over
data_mipa_lulus_over = data_mipa_lulus_over.reset_index(drop=True)
display(data_mipa_lulus_over.head())
```

## Oversampling data sosial

```python
# #### OVERSAMPLING DATA SOSIAL CLASS IPK + LULUS

# #### Convert Pandas to numpy 
attr_sosial_ipk = attr_sosial_ipk.reset_index()
attr_sosial_lulus = attr_sosial_lulus.reset_index()


savedcols = attr_sosial_ipk.columns

attr_sosial_ipk = attr_sosial_ipk.values.astype("U")
attr_sosial_lulus = attr_sosial_lulus.values.astype("U")

class_sosial_ipk = class_sosial_ipk.values.astype("U")
class_sosial_lulus = class_sosial_lulus.values.astype("U")


# # print (attr_sosial_ipk)
# # print (type(attr_sosial_ipk))

# #### oversampling
rus = RandomOverSampler()
data_sosial_ipk_over, class_sosial_ipk_over = rus.fit_sample(attr_sosial_ipk,class_sosial_ipk)
data_sosial_lulus_over, class_sosial_lulus_over = rus.fit_sample(attr_sosial_lulus, class_sosial_lulus)

# # print (data_sosial_ipk_over)
# # print (savedcols)

# # convert back to pandas
data_sosial_ipk_over = pd.DataFrame(data_sosial_ipk_over, columns=savedcols).set_index("index")
data_sosial_ipk_over['IPK_TERAKHIR'] = class_sosial_ipk_over
data_sosial_ipk_over = data_sosial_ipk_over.reset_index(drop=True)
display(data_sosial_ipk_over.head())

data_sosial_lulus_over = pd.DataFrame(data_sosial_lulus_over, columns=savedcols).set_index("index")
data_sosial_lulus_over['STATUS_LULUS']=class_sosial_lulus_over
data_sosial_lulus_over = data_sosial_lulus_over.reset_index(drop=True)
# data_sosial_lulus_over = data_sosial_lulus_over.drop('IPK_TERAKHIR', axis = 1)
display(data_sosial_lulus_over.head())
```
