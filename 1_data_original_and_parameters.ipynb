# Plot Functions

```python

'''
Plot function to display missing value for each Departement
'''
data_plotmissing_dept = widgets.Dropdown(
    options = ['Matematika','Biologi','Fisika', 'Informatika', 
             'Akuntansi','Ekonomi','Komunikasi','Hukum'],
    value = 'Matematika',
    description ='Data:',
    disabled = False
)

chart_plotmissing_dept = widgets.Dropdown(
    options = ['Matrix', 'Bar'],
    value ='Matrix',
    description = 'Chart:',
    disabled = False,
)

def plotmissing_dept(data,chart):
    x = "Grafik "+chart+" "+data
    print(x)
    if data =='Matematika' :
        if chart == "Matrix": return msno.matrix(datafull_mtk)
        if chart == 'Bar': return msno.bar(datafull_mtk)
        
    if data =='Biologi' :
        if chart == "Matrix": return msno.matrix(datafull_bio)
        if chart == 'Bar': return msno.bar(datafull_bio)
        
    if data =='Fisika' :
        if chart == "Matrix": return msno.matrix(datafull_fisika)
        if chart == 'Bar': return msno.bar(datafull_fisika)
        
    if data =='Informatika' :
        if chart == "Matrix":return msno.matrix(datafull_if)
        if chart == 'Bar': return msno.bar(datafull_if)

    if data =='Akuntansi' :
        if chart == "Matrix":return msno.matrix(datafull_akutansi)
        if chart == 'Bar': return msno.bar(datafull_akutansi)
    
    if data =='Ekonomi' :
        if chart == "Matrix":return msno.matrix(datafull_ekonomi)
        if chart == 'Bar': return msno.bar(datafull_ekonomi)

    if data =='HI' :
        if chart == "Matrix":return msno.matrix(datafull_hi)
        if chart == 'Bar': return msno.bar(datafull_hi)

    if data =='Komunikasi' :
        if chart == "Matrix":return msno.matrix(datafull_komunikasi)
        if chart == 'Bar': return msno.bar(datafull_komunikasi)
         
    if data =='Hukum' :
        if chart == "Matrix":return msno.matrix(datafull_hukum)
        if chart == 'Bar': return msno.bar(datafull_hukum)
    

'''
Plot function to display missing value for each Science
'''
data_plotmissing_sci = widgets.Dropdown(
    options=['IPA','SOSIAL'],
    value='IPA',
    description='Data:',
    disabled=False,
)

chart_plotmissing_sci = widgets.Dropdown(
    options=['Matrix', 'Bar'],
    value='Matrix',
    description='Chart:',
    disabled=False,
)

def plotmissing_sci(data,chart):
    x = "Grafik "+chart+" "+data
    print(x)
    
    if data =='IPA' :
        if chart == "Matrix":return msno.matrix(data_mipa)
        if chart == 'Bar': return msno.bar(data_mipa)
    
    if data =='SOSIAL' :
        if chart == "Matrix": return msno.matrix(data_sosial)
        if chart == "Bar": return msno.bar(data_sosial)
        
'''
Plot function to display count sum every parameter in each dataset
''' 

data_attr_original = widgets.Dropdown(
    options = ['IPA','SOSIAL'],
    value ='IPA',
    description ='Data:',
    disabled = False
)

yaxis_attr_original = widgets.Dropdown(
    options = {'Jenis Kelamin':'JK', 
               'Angkatan':'ANGKATAN',
               'Status Mahasiswa':'STATUS_MHS',
               'Pekerjaan Ayah':'PEK_AYAH',
               'Pekerjaan Ibu':'PEK_IBU', 
               'Pendidikan Ibu':'PDK_IBU', 
               'Gaji':'GAJI_ORTU'},
    value = 'JK',
    description = 'Parameter:',
    disabled = False
)

def plot_attr_original(data, yaxis):
    x = "Grafik atribut "+yaxis+" "+data
    print(x)
    
    if data =='IPA' :
        plt.figure(figsize=(10,5))
        sns.countplot(y=yaxis,data=data_mipa, palette="Greens_d")
        #plt.xlim(0, 2100)
        plt.show()
    
    if data =='SOSIAL' :
        plt.figure(figsize=(10,5))
        sns.countplot(y=yaxis,data=data_sosial, palette="Blues_d")
        #plt.xlim(0, 2100)
        plt.show()
    
def color_negative_blue(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = '#c4f4d6' if val == 0 else 'None'
    return 'background-color: %s' % color
                               


```

# Merging GPA & Demographic Data

```python
'''
Merging GPA & DEMOGRAPHIC data for each Departement as one 
'''

datafull_mtk = data_mtk.merge(ipk_mtk,left_on ='NIM', right_on ='NIM', how ='left')
datafull_bio = data_biologi.merge(ipk_biologi, left_on ='NIM', right_on ='NIM', how ='left')
datafull_fisika = data_fisika.merge(ipk_fisika, left_on ='NIM', right_on ='NIM', how ='left')
datafull_if = data_if.merge(ipk_if,left_on ='NIM', right_on ='NIM', how ='left')

datafull_akutansi = data_akutansi.merge(ipk_akutansi,left_on = 'NIM', right_on = 'NIM', how ='left')
datafull_ekonomi = data_ekonomi.merge(ipk_ekonomi,left_on ='NIM', right_on = 'NIM', how ='left')
datafull_komunikasi = data_komunikasi.merge(ipk_komunikasi, left_on ='NIM', right_on = 'NIM', how ='left')
datafull_hukum = data_hukum.merge(ipk_hukum,left_on ='NIM', right_on ='NIM', how ='left')

'''
Merging Data from all departements and divided two such as
Natural Science and Sosial Science
'''

#merge Data Demografis student and Data IP
data_mipa =pd.concat([datafull_fisika, datafull_mtk,datafull_bio,datafull_if],sort=True)
data_mipa = data_mipa.reset_index(drop=True)

data_sosial =pd.concat([datafull_akutansi, datafull_ekonomi,datafull_komunikasi, datafull_hukum], sort=True)
data_sosial = data_sosial.reset_index(drop=True)
data_sosial = data_sosial[data_sosial.columns[~data_sosial.columns.str.contains('Unnamed:')]]

```

# Pemilihan Data Angkatan
Melakukan proses droping/deleting data mahasiswa ilmu sosial (data_sosial) angkatan < 2002 dan angkatan > 2013

```python
# drop/delete row who has value in column/parameter angkatan <2000 or > 2013
data_sosial.drop(data_sosial[(data_sosial.ANGKATAN < 2002) | (data_sosial.ANGKATAN > 2013)].index, inplace=True)


# Reindex data sosial
data_sosial = data_sosial.reset_index(drop=True)


# Drop None value in every columns
mask = data_sosial.applymap(lambda x: x is None)
cols = data_sosial.columns[(mask).any()]
for col in data_sosial[cols]:
    data_sosial.loc[mask[col], col] = ''

    
# Selecting row without null value    
data_sosial= data_sosial[data_sosial['IPK_TERAKHIR'].notnull()]

# data_sosial= data_sosial[data_sosial['IP_S1'].notnull()]
# # data_sosial= data_sosial[data_sosial['IP_S2'].notnull()]
# data_sosial= data_sosial[data_sosial['IP_S3'].notnull()]
# data_sosial= data_sosial[data_sosial['IP_S4'].notnull()]
# data_sosial= data_sosial[data_sosial['IP_S5'].notnull()]
# data_sosial= data_sosial[data_sosial['IP_S6'].notnull()]
# data_sosial= data_sosial[data_sosial['IP_S7'].notnull()]


# data_sosial= data_sosial[data_sosial['SKS_S1'].notnull()]
# data_sosial= data_sosial[data_sosial['SKS_S2'].notnull()]
data_sosial.drop(data_sosial[(data_sosial.SKS_S2 > 24)].index,inplace =True)
data_sosial.drop(data_sosial[(data_sosial.SKS_S1 > 24)].index,inplace =True)
data_sosial.drop(data_sosial[(data_sosial.IP_S2 > 4.00)].index,inplace =True)
data_sosial.drop(data_sosial[(data_sosial.IP_S1 > 4.00)].index,inplace =True)


data_sosial = data_sosial.dropna(thresh = 10)
```
